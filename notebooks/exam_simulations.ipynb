## State modelling

Let's start by modelling the state. We need to track the number of monkey on both sides.

We can model the state by considering only the monkey on the left. 
The monkeys on the right will be the difference between 4 - state.

state = 4

The goal state, then, should be

goal_state = 0

where there's no monkey left on the left side.

However, to model the actions we should know where the torch is! So let's add the torch to the state representation.

initial_state = 4, 'left'
goal_state = 0, 'right'

In this case the action move 2 monkey from the left to the right will be:

action = 2, 'left'

To make the implementation more explicit and the problem easier to code let's explicitly report the number of monkeys on both sides

initial_state = 4, 0, 'left'
goal_state = 0, 4, 'right'

To not deal with strings let's encode 'left' with 0 and 'right' with 1, since the 0-index in the state are the monkeys on the left, and the 1-index are the monkeys on the right side

initial_state = (4, 0 , 0)
goal_state = (0, 4, 1)

## Actions

The actions will be something like

a1 = (1, 0) -> move 1 monkey from left side
a2 = (2, 1) -> move two monkeys from right side

```python
class ForestProblem:

    def __init__(self, initial_state=None, goal_state=None):

        # I need to define the initial state
        self.initial_state = (4, 0, 0)
        self.goal_state = (0, 4, 1)

    def actions(self, state):
        actions = []
        # I can move only monkeys on the same side of the torch
        torch_position = state[2]
        # Number of monkeys on the same side as the torch
        movable_monkeys = state[torch_position]
        # move one monkey if there is at least one monkey on the same side as the torch
        if movable_monkeys > 0:
            actions.append([1, torch_position])
        # move two monkeys if there are at least two monkeys on the same side as the torch
        if movable_monkeys > 1:
            actions.append([2, torch_position])
        return actions
```

## Action result

Given an action we have to remove monkeys from one side and add them to the other side, depending on where the torch is.

```python
class ForestProblem:

    def __init__(self, initial_state=None, goal_state=None):

        # I need to define the initial state
        self.initial_state = (4, 0, 0)
        self.goal_state = (0, 4, 1)

    def actions(self, state):
        actions = []
        # I can move only monkeys on the same side of the torch
        torch_position = state[2]
        # Number of monkeys on the same side as the torch
        movable_monkeys = state[torch_position]
        # move one monkey if there is at least one monkey on the same side as the torch
        if movable_monkeys > 0:
            actions.append([1, torch_position])
        # move two monkeys if there are at least two monkeys on the same side as the torch
        if movable_monkeys > 1:
            actions.append([2, torch_position])
        return actions
    
    def result(self, state, action):
        # extract needed information from the action
        monkeys_to_move = action[0]
        torch_position = action[1]
        # create a new state based on the current state, as a list to be able to modify it
        new_state = list(state)
        # move monkeys from left to right
        if torch_position == 0:
            new_state[0] -= monkeys_to_move
            new_state[1] += monkeys_to_move
            # update torch position
            new_state[2] = 1
        # move monkeys from right to left
        elif torch_position == 1:
            new_state[1] -= monkeys_to_move
            new_state[0] += monkeys_to_move
            # update torch position
            new_state[2] = 0
        else:
            # torch position can be only 0 or 1
            raise ValueError('Wrong torch value')
        # monkeys must always be between 0 and 4 on both sides
        if new_state[0] < 0 or new_state[0] > 4:
            raise ValueError(f'Wrong state: {new_state}')
        if new_state[1] < 0 or new_state[1] > 4:
            raise ValueError(f'Wrong state: {new_state}')
        return tuple(new_state)
```

```python
import random

class DepthLimitedStrategy:

    def select(self, fringe):
        # let's remove from the fringe all nodes that are deeper than the limit
        limit = 10
        fringe = [node for node in fringe if node.depth <= limit]
        # let's follow a depth first approach
        selected_node = fringe.pop()
        return fringe, selected_node
```

## Let's modify the search so it behaves like a Tree Search

We need to remove the comparison with the already visited states

```python
from path_search.search import Node

def expand(node, problem):
    actions = problem.actions(node.state)
    new_nodes = []
    for action in actions:
        new_state = problem.result(node.state, action)
        new_node = Node(state=new_state, action=action, parent=node, path_cost=node.path_cost+problem.action_cost(node.state, action))    
        new_nodes.append(new_node)
    return new_nodes

class Search:

    def __init__(self, problem, strategy) -> None:
        
        self.problem = problem
        self.strategy = strategy

    def run(self):
        root_node = Node(state=self.problem.initial_state, action=None, parent=None, path_cost=0, depth=0)
        fringe = [root_node]
        visited = dict()

        while len(fringe) > 0:
            fringe, node = self.strategy.select(fringe)
            # it is sufficient to comment the code lines
            # if node.state in visited and visited[node.state] <= node.path_cost:
            #     continue

            if self.problem.is_goal(node.state):
                return node

            visited[node.state] = node.path_cost
            new_nodes = expand(node, self.problem)
            fringe = fringe + new_nodes
            # input('Press Enter to continue...')
        
        return None


```

## Let's now design an heuristic

An heuristic can be derived by considering a simplified version of the same problem. An heuristic is a cost estimation.

The simplified problem can be a problem without the torch constraint. So the cost is the number of moves needed to move the monkeys without the torch requirement.

```python
class ForestProblem:

    def __init__(self, initial_state=None, goal_state=None):

        # I need to define the initial state
        self.initial_state = (4, 0, 0)
        self.goal_state = (0, 4, 1)

    def actions(self, state):
        actions = []
        # I can move only monkeys on the same side of the torch
        torch_position = state[2]
        # Number of monkeys on the same side as the torch
        movable_monkeys = state[torch_position]
        # move one monkey if there is at least one monkey on the same side as the torch
        if movable_monkeys > 0:
            actions.append([1, torch_position])
        # move two monkeys if there are at least two monkeys on the same side as the torch
        if movable_monkeys > 1:
            actions.append([2, torch_position])
        return actions
    
    def result(self, state, action):
        # extract needed information from the action
        monkeys_to_move = action[0]
        torch_position = action[1]
        # create a new state based on the current state, as a list to be able to modify it
        new_state = list(state)
        # move monkeys from left to right
        if torch_position == 0:
            new_state[0] -= monkeys_to_move
            new_state[1] += monkeys_to_move
            # update torch position
            new_state[2] = 1
        # move monkeys from right to left
        elif torch_position == 1:
            new_state[1] -= monkeys_to_move
            new_state[0] += monkeys_to_move
            # update torch position
            new_state[2] = 0
        else:
            # torch position can be only 0 or 1
            raise ValueError('Wrong torch value')
        # monkeys must always be between 0 and 4 on both sides
        if new_state[0] < 0 or new_state[0] > 4:
            raise ValueError(f'Wrong state: {new_state}')
        if new_state[1] < 0 or new_state[1] > 4:
            raise ValueError(f'Wrong state: {new_state}')
        return tuple(new_state)
  
    def action_cost(self, state, action):
            return 1
    
    def is_goal(self, state):
        return state == self.goal_state
    
    def heuristic(self, state):
        # monkeys on the left
        monkeys = state[0]
        if monkeys == 4 or monkeys == 3:
            return 2
        elif monkeys == 2 or monkeys == 1:
            return 1
        elif monkeys == 0:
            return 0
        else:
            raise ValueError(f'Wrong state: {state}')
```

```python
class AStar:

    def __init__(self, problem) -> None:
        self.problem = problem

    def select(self, fringe):
        fringe.sort(key=lambda node: self.problem.heuristic(node.state) + node.path_cost)
        return fringe, fringe.pop(0)
```

## Actions 

We need to ranomly select 20 neighbouring states. A neighbouring state is a state where we replace one player with another.

```python
class BariPlayersProblem:
    def __init__(self):
        self.initial_state = self.random_state()

    def random_state(self):
        # we need to generate a random state with 11 players, each of them can be any of the players in the list
        import random
        random_state = []
        # loop 11 times to generate a random state with 11 players
        for _ in range(11):
            random_state.append(random.choice(players_names))
        return tuple(random_state)
    

    def actions_(self, state):
        # action: remove player1 and insert player 2
        actions = []
        players_in_state = list(state)
        # we need to generate 20 random actions a.k.a. 20 random neighbors of the current state
        for _ in range(20):
            # select a random player to remove from the state
            p1 = random.choice(players_in_state)
            # select a random player to insert in the state, but it must not be already in the state
            done = False
            while not done:
                p2 = random.choice(players_names)
                # if p2 is not in the state, we can insert it
                if p2 not in players_in_state:
                    done = True
            actions.append((p1, p2))
        return actions
    
```

## Evaluation

In local search the most important part is the performance measure definition. 

The value is given by the summation of the players values.

To impose constraints we add penalties greather than the maximum team value. 

```python
PENALTY = 100

class BariPlayersProblem:
    def __init__(self):
        self.initial_state = self.random_state()

    def random_state(self):
        import random
        random_state = []
        for _ in range(11):
            random_state.append(random.choice(players_names))
        return tuple(random_state)

    def actions(self, state):
        # action: remove player1 insert player 2
        actions = []
        players_in_state = list(state)
        for _ in range(20):
            # player to remove
            p1 = random.choice(players_in_state)
            # player to insert
            done = False
            while not done:
                p2 = random.choice(players_names)
                if p2 not in players_in_state:
                    done = True
            actions.append((p1, p2))
        return actions


    def result(self, state, action):
        state = list(state)
        p1, p2 = action
        state.remove(p1)
        state.append(p2)
        return tuple(state)

    def count_role(self, state, role):
        count = 0
        for p in state:
            if players[p]['ruolo'] == role:
                count += 1
        return count
    
    def team_value(self, state):
        value = 0
        for p in state:
            value += players[p]['valore']
        return value

    def evaluate(self, state):
        value = self.team_value(state)
        # constraints penalties
        if self.count_role(state, 'Portiere') == 0:
            value -= PENALTY
        if self.count_role(state, 'Difensore') < 3:
            value -= PENALTY
        if self.count_role(state, 'Centrocampista') < 2:
            value -= PENALTY
        if self.count_role(state, 'Attaccante') == 0:
            value -= PENALTY

        return value
```
